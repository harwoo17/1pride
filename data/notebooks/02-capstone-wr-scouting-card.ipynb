## Section 1 — Pull the data

One query pulls every NFC North receiver-week. Aggregations happen in pandas so you can experiment with metrics without re-querying.

```python
raw = pd.read_sql(
    """
    SELECT player_display_name, recent_team, position_group, week,
           targets, receptions, receiving_yards, receiving_tds, receiving_epa
    FROM weekly_stats
    WHERE season = 2024
      AND season_type = 'REG'
      AND recent_team = ANY(%(teams)s)
      AND position_group IN ('WR', 'TE')
      AND targets IS NOT NULL
      AND targets > 0
    """,
    eng,
    params={'teams': list(DIVISION)},
)
raw.head()
```
